```python
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.nodeName = defaultdict(list)

    def createNode(self, node, pointed_node, name=''):
        self.graph[node].append(pointed_node)
        self.nodeName[name].append(node)

    def breadth(self, root):
        visited = [False] * (len(self.graph))

        queue = []

        queue.append(root)
        visited[root] = True
        i = 1
        while queue:


            root = queue.pop(0)
            print ("Node ", self.nodeName[root]," N_o ",root)
            
            i = i + 1
            
            for i in self.graph[root]:
                if visited[i] == False:
                    queue.append(i)
                    visited[i] = True
```
